```python

import matplotlib.pyplot as plt
import cv2
```

```python
	##import and read image which we want to convert to sketch
image = cv2.imread('/content/input2-300x295.jpg')
```

```python
##image in the form of array(RBG IMAGE)
image
```

```python
#showing RBG image
plt.imshow(image)
```

```python
##converting image to graystyle	
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
##gray image
plt.imshow(gray)
```

```python
##inverting gray image
img_not = cv2.bitwise_not(gray)
```

```python
plt.imshow(img_not)
```

```python
##making inverted that is negative gray image blurr
blur = cv2.GaussianBlur(img_not, (21, 21), 0)
```

```python
invertedblur = cv2.bitwise_not(blur)
```

```python
##making pencil sketch by combining gray and invertedblurr image
pencil_sketch = cv2.divide(gray, invertedblur, scale=256.0)
```

```python
pencil_sketch
```

```python
plt.imshow(pencil_sketch)
```
